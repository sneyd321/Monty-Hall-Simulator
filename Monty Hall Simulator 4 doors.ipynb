```python
#Assignment 2 for Statistics for Data Science
#MATH37198 - Ryan Sneyd
import numpy as np
import pandas as pd
from random import randint

#initialize lists for data entry
prizeList = []
initialList = []
revealedList1 = []
revealedList2 = []
finalList = []

#add numbers to lists 10000 times
for i in range(0, 10000):
    #Create random values for each list
    randomPrize = randint(1, 4)
    randomInitial = randint(1, 4)
    randomRevealed1 = randint(1, 4)
    randomRevealed2 = randint(1, 4)
    randomFinal = randint(1, 4)
    #Check to make sure all values in table are different
    stopCheck = True
    while (stopCheck):
        #Make sure the doors revealed1 is different from initial and final
        if  randomRevealed1 == randomInitial or randomRevealed1 == randomFinal:
            #if true make a new revealed1 value
            randomRevealed1 = randint(1, 4)
        #Make sure the doors revealed2 is different from initial and final
        elif  randomRevealed2 == randomInitial or randomRevealed2 == randomFinal:
            #if true make a new revealed2 value
            randomRevealed2 = randint(1, 4)
        #Make sure revealed1 is different from revealed2
        elif randomRevealed1 == randomRevealed2:
            #if true make a new revealed1 value
            randomRevealed1 = randint(1, 4)
            #if true make a new revealed2 value
            randomRevealed2 = randint(1, 4)
        else:
            #Finally make sure the prize does not equal revealed
            if randomPrize == randomRevealed1 or randomPrize == randomRevealed2:
                #if true make a new prize value
                randomPrize = randint(1, 4)
            #If everything works break out of loop
            else:
                stopCheck = False
    
    
    #add all values to lists
    prizeList.append(randomPrize)
    initialList.append(randomInitial)
    revealedList1.append(randomRevealed1)
    revealedList2.append(randomRevealed2)
    finalList.append(randomFinal)
#convert list in nparray beacues it is easier to insert into dataframe
pl = np.asarray(prizeList)   
il = np.asarray(initialList)  
rl1 = np.asarray(revealedList1)  
rl2 = np.asarray(revealedList2)  
fl = np.asarray(finalList)  
#create dataframe and add values
data = pd.DataFrame({"Prize" : pl, "Initial" : il, "Revealed1" : rl1, "Revelaed2" : rl2, "Final" : fl})
#output data
data

```

```python
#What is the probability of a contestant winning given the contestant always switches doors
#Set a variable to count where the contestant switched doors
switchedCounter = 0
#Set a variable to count where the switches and wins
winCounter = 0
#For every vale in the table
for i in range(1, len(data)):
    #If the contestant switches and wins
    if data["Initial"].values.item(i) != data["Final"].values.item(i) and data["Prize"].values.item(i) == data["Final"].values.item(i):
        winCounter += 1
        #If the contestant switches 
    if data["Prize"].values.item(i) != data["Final"].values.item(i):
        switchedCounter += 1
        
#Total number of time switched
total = winCounter/switchedCounter

#Output
print("The percentage of trials the contestant wins by switching doors is: " + str(round((total)*100, 2))+"%")
```
